Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


Ans: Seaborn is a powerful data visualization library built on top of Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Here are five common types of plots you can create using Seaborn, along with their uses:

Scatter Plot: To explore relationships between two continuous variables.

Line Plot: To visualize trends over time or ordered categories.

Bar Plot: To compare categories based on mean values.

Heatmap: To visualize correlation or intensity of data in a matrix.

Box Plot: To compare distributions and detect outliers in datasets.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

df_fmri = sns.load_dataset('fmri')

plt.figure(figsize=(6,2))
sns.lineplot(x='timepoint', y='signal', hue='event', style='region', data=df_fmri)
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.title('Line Plot of Signal Over Time for Different Events and Regions')
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.


```python
titanic_df = sns.load_dataset('titanic')

plt.figure(figsize=(6,2))

plt.subplot(1,2,1)
sns.boxplot(x='pclass', y='age', data=titanic_df)
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title('Box Plot of Age by Pclass')

plt.subplot(1,2,2)
sns.boxplot(x='pclass', y='fare', data=titanic_df)
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Box Plot of Fare by Pclass')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.


```python
df = sns.load_dataset('diamonds')

plt.figure(figsize=(6,2))
sns.histplot(data=df, x='price', hue='cut', kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Diamond Prices by Cut')
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.


```python
iris_df = sns.load_dataset('iris')

sns.pairplot(iris_df,hue="species")
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights_df = sns.load_dataset('flights')

flights_df = flights_df.pivot(index='month', columns='year', values='passengers')

plt.figure(figsize=(6,2))
sns.heatmap(flights_df, annot=True, fmt='d', cmap='YlGnBu')
```
